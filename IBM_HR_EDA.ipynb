```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('IBM.csv')
df.head()
```

```python
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
#Attrition Count plot
sns.countplot(x='Attrition', data=df)
plt.show()
print()
# Attrition percentage
attrition_rate = df['Attrition'].value_counts(normalize=True) * 100
print(attrition_rate)
```

### Insight: Around 16% of employees leave the company, indicating moderate attrition.

```python
# Salary vs Attrition
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title("Salary vs Attrition")
plt.show()
```

### Insight: Employees who leave generally have lower salaries, suggesting pay is a key factor.

```python
#department vs attrition
sns.countplot(x='Department', hue='Attrition', data=df)
plt.xticks(rotation=45)
plt.show()
```

### Insight: Sales department show higher attrition, indicating department-specific issues.

```python
#Job Role impact
plt.figure(figsize=(10,5))
sns.countplot(y='JobRole', hue='Attrition', data=df)
plt.title("Job Role vs Attrition")
plt.show()
```

### Insight: Certain job roles have more employees leaving, suggesting role-related stress or dissatisfaction.

```python
#overtime impact
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.show()
```

### Insight: Employees who work overtime are more likely to leave, indicating workload pressure.

```python
#age distribution
sns.histplot(df['Age'], kde=True)
plt.show()
```

### Insight: Most employees fall in a specific age range, which is useful for workforce planning.

```python
sns.countplot(x='WorkLifeBalance', hue='Attrition', data=df)
plt.title("Work-Life Balance vs Attrition")
plt.show()
```

### Insight: While poor work-life balance is a clear driver of higher attrition rates, achieving an ideal balance reduces turnover risk but doesn't eliminate it entirely, as other factors still influence the decision to leave.

```python
sns.countplot(x='JobSatisfaction', hue='Attrition', data=df)
plt.title("Job Satisfaction vs Attrition")
plt.show()
```

### Insight: Higher job satisfaction clearly correlates with better retention, but it isn't a total guarantee against turnover.

```python
#Experience vs Attrition
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title("Years at Company vs Attrition")
plt.show()
```

### Insight: Employees leave more in their early years, indicating early dissatisfaction.

```python
#Promotional Gap vs Attrition
sns.boxplot(x='Attrition', y='YearsSinceLastPromotion', data=df)
plt.title("Promotion Gap vs Attrition")
plt.show()
```

### Insight: Most people who leave do so within 0 to 2 years of their last promotion.

```python
# Monthly Income distribution
sns.histplot(df['MonthlyIncome'], kde=True)
plt.title("Monthly Income Distribution")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(14,10))
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# ==========
# Conclusion
# ==========

# This analysis shows that employee attrition (~16%) is influenced by multiple factors.
# Employees who work overtime, have lower salaries, and poor work-life balance are more likely to leave.
# Attrition is also higher among employees in their early years and those who have not been promoted for a long time.
# Certain departments and job roles contribute more to attrition in terms of count.
# These insights suggest that improving salary structure, reducing workload, and providing better growth opportunities can help reduce employee turnover.
